```python
def instructions_game():
    introduction_game = "WELCOME TO TIC TAC TOE game!"
    rule1 = "The game is played on a grid that's 3 squares by 3 squares."
    rule2 = "There are two players known as X and O."
    rule3 = "Players alternate placing Xs and Os on the game board"
    rule4 = "The first player to get 3 of her marks in a row (up, down, across, or diagonally) is the winner."
    rule5 = "When all 9 squares are full, the game is over (no one has three in a row)."
    
    return print(introduction_game+ "\n" + rule1 +"\n" + rule2 +"\n"+rule3 +"\n"+ rule4 +"\n"+ rule5 + "\n")
```

```python
import time
import datetime


# Create function that acts as a countdown
def countdown(h, m, s):
    # Calculate the total number of seconds
    total_seconds = h * 3600 + m * 60 + s
    # While loop that checks if total_seconds reaches zero
    # If not zero, decrement total time by one second
    while total_seconds > 0:
        # Timer represents time left on countdown
        timer = datetime.timedelta(seconds = total_seconds)
        # Prints the time left on the timer
        print(timer, end="\r")
        # Delays the program one second
        time.sleep(1)
        # Reduces total time by one second
        total_seconds -= 1
    print("Bzzzt! The countdown is at zero seconds!")

    
h = input("Enter the time in hours: ")
m = input("Enter the time in minutes: ")
s = input("Enter the time in seconds: ")
countdown(int(h), int(m), int(s))
```

```python
from math import inf as infinity


#First step is to make a board. List of all the positions
game_board = ["*", "*", "*", "*", "*", "*", "*", "*", "*"]


def show_board(game_board):
    print(" ", game_board[0], "|", game_board[1], "|", game_board[2], " ")
    print("----+---+----")
    print(" ", game_board[3], "|", game_board[4], "|", game_board[5], " ")
    print("----+---+----")
    print(" ", game_board[6], "|", game_board[7], "|", game_board[8], " ")
    

#Evaluation of state. param state: the state of the current board.
#return: +1 if the computer wins; -1 if the human wins; 0 draw. returns the score.  
def evaluation(state):
    #winner is robot.
    if winner(state, "o"):
        score = +1
    #winner is human player.
    elif winner(state, "x"):
        score = -1
    else:
        #score for draw
        score = 0      
    return score


#Each empty cell will be added into cells' list and return a list of empty cells.
#i = index, j = empty or not variable
def available_spots(game_board):
    #empty list
    result = []
    for i, j in enumerate(game_board):
        #checks if variable is empty place
        if j == "*":
            result.append(i)          
    return result


"""
This function tests if a specific player wins. Possibilities:
* Three rows    [X X X] or [O O O]
* Three cols    [X X X] or [O O O]
* Two diagonals [X X X] or [O O O]
:param state: the state of the current board
:param player: a human or a computer
:return: True if someone win
"""
def winner(state, player):
    win_states = [
        [state[0], state[1], state[2]],
        [state[3], state[4], state[5]],
        [state[6], state[7], state[8]],
        [state[0], state[3], state[6]],
        [state[1], state[4], state[7]],
        [state[2], state[5], state[8]],
        [state[0], state[4], state[8]],
        [state[2], state[4], state[6]],
    ]
    #player can be human or robot
    if [player, player, player] in win_states:
        return True
    else:
        return False
    
    
#param state: the state of the current board. return: True if the human or computer wins. 
def end_of_game(state):
    return winner(state, "x") or winner(state, "o")


#Cover all the positions in the board because passes aver all free cells and checks for best move.
#Recursive function. 
def minimax(game_board, depth, player):
    #checks if turn of robot (itself)
    if player == "o":
        #best move, puts the smallest. -1 = move and -infinity = score.
        best = [-1, -infinity]
    else:
        best = [-1, infinity]
    
    if depth == 0 or end_of_game(game_board):
        score = evaluation(game_board)
        #-1 = no place and score is for now, before check all cells.
        return [-1, score]
    
    #check every cell in available spots (function that returns an index).
    for cell in available_spots(game_board):
        game_board[cell] = player   
        #call again minimax for each "x" and "o" until game ends.
        if player == "o":
            score = minimax(game_board, depth-1, "x")
        else:
            score = minimax(game_board, depth-1, "o")
        #return the board place to nothing ("*"), after knowing score.
        game_board[cell] = "*"
        #score in the first place. not only -1 score fro place checked.
        score[0] = cell
        if player == "o":
            #minimize
            if best[1] < score[1]:
                best = score
        #turn of "x"
        else:
            if best[1] > score[1]:
                best = score    
    return best

    
#The Human plays choosing a valid move.
def human_turn(game_board):
    #depth is the number of available spots.
    depth = len(available_spots(game_board))
    if depth == 0 or end_of_game(game_board):
        #continue the main function because return = exit.
        return
    #means no index. Nothing!
    move = -1
    #NOT VALID cell in the game board
    while move < 1 or move > 9:
        print("Human player turn\n")
        show_board(game_board)
        try:
            move = int(input("Enter choice 1 to 9: "))
        except ValueError:
            print("Integer only...")
        #VALID cell in the game board
        if move <= 9 and move >= 1:
            #indexes in python starts from 0, not 1.
            if game_board[move-1] == "*":
                move -= 1
                #prints "x" in the selected position of the board (1 to 9).
                game_board[move] = "x"
                show_board(game_board)
                return
            else:
                print("Wrong choice! Position ocupied")
                #means move to nothing
                move =-1
        else:
            print("Wrong choice")
            move = -1
            
def robot_move():
    #depth is all the available spots. Number of available spots.
    depth = len(available_spots(game_board))
    #if no available spots or game over, return = exit loop and continue in the main function.
    if depth == 0 or end_of_game(game_board):
        #exit function and continue in the main function.
        return
    print("Robot turn \n")
    move = minimax(game_board, depth, "o")
    #gets the best position (move previous line) and put "o" in the first best place  
    game_board[move[0]] = "o"
    show_board(game_board)
    countdown(int(h), int(m), int(s))


def main(game_board):
    #loop to check for available spots in the board and game is NOT over
    while len(available_spots(game_board)) > 0 and not end_of_game(game_board):
        #first to move in the game is the human
        human_turn(game_board)
        #second to move in the game
        robot_move()
    #checking for winner after previous while loop ends. Human never wins! This check is false always.
    if winner(game_board, "x"):
        print("Human winner!")
        return 0
    elif winner(game_board, "o"):
        print("Robot winner!")
        return 0
    else:
        print("It is a draw")
        return 0        
```
